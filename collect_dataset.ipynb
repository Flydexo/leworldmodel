```python
from pathlib import Path
import os
import stable_worldmodel as swm
from stable_worldmodel.envs.two_room import ExpertPolicy
```

```python
root = Path(os.environ.get('STABLEWM_HOME', Path.home() / '.stable_worldmodel'))
dataset_path = root / 'datasets' / 'tworooms.lance'

world = swm.World(
    'swm/TwoRoom-v1',
    num_envs=4,
    image_shape=(64, 64)
)

world.set_policy(ExpertPolicy())

world.collect(
    dataset_path,
    episodes=256,
    seed=0,
    options={
        'variation': [
            'agent.position',
            'target.position',
            'door.position'
        ],
    },
)

world.close()
print(f'wrote {dataset_path}')
```
