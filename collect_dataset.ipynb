```python
from pathlib import Path
import os
import stable_worldmodel as swm
from stable_worldmodel.envs.two_room import ExpertPolicy
import random
```

```python
seed = random.randint(0, 10000)
```

```python
root = Path(os.environ.get('STABLEWM_HOME', Path.home() / '.stable_worldmodel'))
dataset_path = root / 'datasets' / 'tworooms-fov.lance'

world = swm.World(
    'swm/TwoRoom-v1',
    num_envs=os.cpu_count(),
    image_shape=(224, 224)
)

world.set_policy(ExpertPolicy())

world.collect(
    dataset_path,
    episodes=10000,
    seed=seed,
    options={
        'variation': [
            'agent.color',
            'agent.position',
            'target.color',
            'target.position',
            'wall.color',
            'wall.thickness',
            'wall.axis',
            'wall.border_color',
            'door.color',
            'door.number',
            'door.size',
            'door.position',
            'background.color',
            'task.min_steps'
        ]
    }
)

world.close()
print(f'wrote {dataset_path}')
```
