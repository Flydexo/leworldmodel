```python
from pathlib import Path
import os
import stable_worldmodel as swm
from stable_worldmodel.envs.two_room import ExpertPolicy
```

```python
root = Path(os.environ.get('STABLEWM_HOME', Path.home() / '.stable_worldmodel'))
dataset_path = root / 'datasets' / 'tworooms-big.lance'

world = swm.World(
    'swm/TwoRoom-v1',
    num_envs=32,
    image_shape=(224, 224)
)

world.set_policy(ExpertPolicy())

world.collect(
    dataset_path,
    episodes=1000,
    seed=0
)

world.close()
print(f'wrote {dataset_path}')
```
